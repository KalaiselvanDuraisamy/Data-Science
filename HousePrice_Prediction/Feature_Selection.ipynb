```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
features = correlation(train_df, 0.8)
features
```

```python
for feature in features:
    train_df = train_df.drop(labels=[feature], axis=1)
```

```python
train_df.head()
```

```python
train_df.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_df, test_x, test_size=0.2, random_state=0)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import SelectPercentile

selected_top_columns = SelectPercentile(mutual_info_regression, percentile=40)
selected_top_columns.fit(X_train, y_train)
features = X_train.columns[selected_top_columns.get_support()]

features
```

```python
train_x = train_df[features]

test_df = pd.read_csv("new_test.csv")
train_y = test_df[features]

train_x.shape, train_y.shape
```
